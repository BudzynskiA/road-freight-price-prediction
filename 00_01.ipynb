```python
# 1. Import core libraries
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# 2. Define path to the Excel file (same folder as the notebook)
DATA_PATH = Path("data_train.xlsx")  # relative path

# 3. Load the workbook (first sheet by default)
df = pd.read_excel(DATA_PATH)

# 4. Inspect basic structure
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nHead:\n", df.head())


# 6. Visual sanity‑check of the EUR price series
if "EUR" in df.columns:
    df["EUR"].plot(title="EUR Price – Raw Series")
    plt.xlabel("Index / Date")
    plt.ylabel("EUR value")
    plt.show()
else:
    print("⚠️  Column 'EUR' not found in the dataset.")
```
