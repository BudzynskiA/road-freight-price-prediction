```python
import pandas as pd
import matplotlib.pyplot as plt

# Load logbook data from CSV
log_df = pd.read_csv('logbook.csv')

# Plot average and best MAPE per generation
plt.figure(figsize=(12, 6))
plt.plot(log_df['gen'], log_df['avg'], label='Average MAPE', marker='o')
plt.plot(log_df['gen'], log_df['min'], label='Best MAPE', marker='s', linestyle='--')

# Add labels and title
plt.xlabel('Generation', fontsize=14)
plt.ylabel('MAPE', fontsize=14)
plt.title('Evolution of MAPE across generations', fontsize=16)

# Add grid and legend
plt.grid(True)
plt.legend(fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
